```python
medals_url = "http://www.winterolympicsmedals.com/medals.csv"
```

```python
def sistem_donwload(url,filename):
    import os
    import urllib3
    http = urllib3.PoolManager()
    r = http.request('Get',url)
    r.status

    response = r.data
    str_data = response.decode("utf-8")

    lines = str_data.split("\n")
    name_cols = lines[0].split(",")
    n_cols = len(cols)
    counter=0
    data_min={}

    for col in  name_cols:
        data_min[col] = []

    for line in lines:
        if(counter > 0):
            values = line.strip().split(",")
            for i in range(len(name_cols)):
                data_min[name_cols[i]].append(values[i])
        counter += 1   

    print("el numero de columnas %d  y el numero de filas -  %d"%(counter,n_cols)) 

    medals_df = pd.DataFrame(data_min)
    print(medals_df.head())

    mainpath ="/home/sonia/Documents/python-ml-course/datasets/"
    #filename ="atletas/atletas_donwload"
    fullpath = os.path.join(mainpath,filename)

    medals_df.to_csv(fullpath + ".csv")
    medals_df.to_json(fullpath + ".json")
    medals_df.to_excel(fullpath + ".xls")
    
    return medals_df
```
